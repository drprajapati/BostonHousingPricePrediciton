```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../input/boston_housing.csv')
dataset.head()
```

```python
features = dataset.loc[:,['crim','zn','nox','rm']].values
labels = dataset['medv'].values
```

```python
plt.scatter(features[:,-1], labels, color="green")
plt.show()
```

```python
features = np.column_stack((np.ones(len(labels)), features))
print(features[:10])
print(features.shape)
```

```python
theta = np.dot(np.linalg.inv(np.dot(np.transpose(features), features)), np.dot(np.transpose(features), labels))
print(theta)

theta0, theta1 = theta[0], theta[4]
```

```python
plt.scatter(features[:,-1], labels, color="green")
plt.plot(features[:,-1], theta0 + theta1 * features[:,-1], color="red")
plt.xlabel('Sq.ft. living')
plt.ylabel('Price in $')
plt.title('Sq.ft. v/s Price')
plt.show()
```
